Bar Plot : This plot allows for comparison of score distributions between male and female students, highlight's differences in performance and identify any gender-based disparities.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/StudentsPerformance.csv')

# Calculate average scores by gender
gender_scores = data.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()

# Reshape the data for easier plotting
gender_scores_melted = gender_scores.melt(id_vars='gender', var_name='Subject', value_name='Score')

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Subject', y='Score', hue='gender', data=gender_scores_melted, palette='Set2')
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xlabel('Subject')
plt.show()
```

Histogram plot: This visualization helps in understanding the distribution of student scores.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/StudentsPerformance.csv')

# Set a style for the plots
sns.set(style="whitegrid")

# Create a figure with multiple subplots
plt.figure(figsize=(18, 6))

# Histogram for Math Scores
plt.subplot(1, 3, 1)
sns.histplot(data['math score'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Math Scores')  # Title for the plot
plt.xlabel('Math Score')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label

# Histogram for Reading Scores
plt.subplot(1, 3, 2)
sns.histplot(data['reading score'], bins=15, kde=True, color='lightgreen')
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')

# Histogram for Writing Scores
plt.subplot(1, 3, 3)
sns.histplot(data['writing score'], bins=15, kde=True, color='salmon')
plt.title('Distribution of Writing Scores')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')


plt.tight_layout()

# Display the plots
plt.show()
```

Scatter Plot: This scatter plot can show the relationship between the different scores compare to math score.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/StudentsPerformance.csv')

# Set style for the plots (optional for better visuals)
sns.set(style="whitegrid")

# Create a figure with two subplots (2 rows, 1 column)
plt.figure(figsize=(12, 10))

# Scatter plot for Math vs. Reading Scores
plt.subplot(2, 1, 1)
sns.scatterplot(x='math score', y='reading score', hue='gender', data=data, palette='Set2')
plt.title('Math Score vs. Reading Score')  # Title for the plot
plt.xlabel('Math Score')  # X-axis label
plt.ylabel('Reading Score')  # Y-axis label

# Scatter plot for Math vs. Writing Scores
plt.subplot(2, 1, 2)
sns.scatterplot(x='math score', y='writing score', hue='gender', data=data, palette='Set2')
plt.title('Math Score vs. Writing Score')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')

plt.tight_layout()

# Display the plots
plt.show()
```

Box plot: this plot visualize diffrent scores with gender.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/StudentsPerformance.csv')

# Set a style for the plots (optional)
sns.set(style="whitegrid")


plt.figure(figsize=(18, 6))

# Box plot for Math Scores by Gender
plt.subplot(1, 3, 1)
sns.boxplot(x='gender', y='math score', data=data, palette='Set2')
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')

# Box plot for Reading Scores by Gender
plt.subplot(1, 3, 2)
sns.boxplot(x='gender', y='reading score', data=data, palette='Set2')
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')

# Box plot for Writing Scores by Gender
plt.subplot(1, 3, 3)
sns.boxplot(x='gender', y='writing score', data=data, palette='Set2')
plt.title('Writing Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Writing Score')

plt.tight_layout()

# Display the plots
plt.show()
```

 Heatmap of Correlation Matrix: correlation of different scores.Reveals relationships between math, reading, and writing scores. Strong correlations suggest that high performance in one subject may predict high performance in another.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/StudentsPerformance.csv')


corr_matrix = data[['math score', 'reading score', 'writing score']].corr()


plt.figure(figsize=(8, 6))

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)

# Add a title
plt.title('Correlation Heatmap of Math, Reading, and Writing Scores')

# Display the heatmap
plt.show()
```
